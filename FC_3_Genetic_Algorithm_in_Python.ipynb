This function creates a hashmap from the adjacency matrix in which the stations with which each station has adjacency are shown in a list, this beacause when using a random election of the next station it only appears the possible arrival stations from the departure stations.

```python
#creation of adjacency hash map
def create_adjacency_hash_map(matrix):
    hash_map = {}
    for i, row in enumerate(matrix):
        connected_nodes = [j for j, value in enumerate(row) if value != float('inf')]
        hash_map[i] = connected_nodes
    return hash_map


adjacency_hash_map = create_adjacency_hash_map(adjacency_matrix)
print(adjacency_hash_map)
```

## Genetic Algorithm functions and variables
In this section it will be presented the functions and variables that the Genetic Algorithm will use.

```python
#variables and function declarations for the GA
Genome = List[int]
Population = List[Genome]
PopulateFunc = Callable[[], Population]
FitnessFunc = Callable[[Genome], int]
SelectionFunc = Callable[[Population, FitnessFunc], Tuple[Genome, Genome]]
CrossoverFunc = Callable[[Genome, Genome], Tuple[Genome, Genome]]
MutationFunc = Callable[[Genome], Genome]
PrinterFunc = Callable[[Population, int, FitnessFunc], None]
```

The first function creates a random genome based in the adjacency matrix and a specified length. The funcion creates a genome that composes of the following structure |station 0|station 1|...|

```python
def generate_genome(length: int) -> Genome:
    #call recursive function that returns the genome, the parameter will be a departure station and it will
    #choose a random possible go station until it chooses a station it have already been in, filling the remaining
    #genome with a specified integer
    departure_station = 0
    explored_stations = [0]
    current_genome = []
    genome = selectRoute(departure_station, explored_stations, current_genome,length)
    #fill with -1 when the actual route ends so all the genomes are the same size
    while len(genome) < length:
        genome.append(-1)
    return genome

def selectRoute(departure_station: int, explored_stations, current_genome, length: int):
    #add the departure station to the genome
    current_genome.append(departure_station)
    #get value of departureStation
    possible_Stations = adjacency_hash_map[departure_station]
    #from that station select a random arrival station
    next_station = choices(possible_Stations)[0]
    #if not explored call recursive function
    if (next_station not in explored_stations) and (len(current_genome)!=length):
        explored_stations.append(next_station)
        selectRoute(next_station, explored_stations,current_genome, length)
    #else return genome at the moment
    return current_genome
```

Probably the most important function for the GA is the fitness function which will indicate the fitness of the genome based on the problem. In this case the main factor was the minimum distance from the station to San Lázaro. In case that the genome have already arrived to San Lázaro the next criteria of evaluation will be the traversed distance of the genoma.

```python
def FitnessFunc(genome: Genome):
    i = 0
    #i get to the last station
    while genome[i+1] != -1 and i<len(genome)-2:
        i += 1
    #get fitness based on the las station distance to San Lázaro
    fitness = minimum_distances_per_station[genome[i]]+100
    #distance criteria if arrived to San Lázaro
    if fitness==100:
      fitness-=(100-(calculate_weight(genome)))
    return fitness
#get the sumatory of the distance of the path traversed by the genome
def calculate_weight(genome):
    total_weight = 0
    for i in range(len(genome) - 1):
        source_node = genome[i]
        target_node = genome[i + 1]
        if target_node != -1:
            weight = adjacency_matrix[source_node][target_node]
            total_weight += weight
        else:
            return total_weight

    return total_weight
#get minimum distance of several genomes in case of tie of fitness
def find_minimum_weight_genome(genomes):
    min_weight = 10000
    min_weight_genome = None
    min_weight_genome2 = None
    for genome in genomes:
        weight = calculate_weight(genome)
        if weight <= min_weight:
            min_weight_genome2 = min_weight_genome
            min_weight = weight
            min_weight_genome = genome

    return min_weight_genome,min_weight_genome2
```

Generates a population of n genomes

```python
#generate size genomes of genome_length size
def generate_population(size: int, genome_length: int) -> Population:
    return [generate_genome(genome_length) for _ in range(size)]
```

GA have to modify and manipulate the genomes in order to explore new possibilities, therefore there are some functions that can allow this. For example the next one cuts two genomes at exactly the same gene and combines them.

```python
def single_point_crossover(a: Genome, b: Genome) -> Tuple[Genome, Genome]:

    if len(a) != len(b):
        print(a)
        print(b)
        raise ValueError("Genomes a and b must be of same length")

    #traverse genome a, save their stations into a hashmap as keys and as values the index in which they locate
    hashMapOfa = {}
    matchingGens = {}
    for i in range(len(a)):
        if a[i]==-1:
            break
        if a[i] == 0:
            continue
        hashMapOfa[a[i]] = i
    #traverse genome b chcecking if the gen is in the already saved hashmap of genome a, if so add to the hashmap of match genes the gene and respective indexes
    for i in range(len(b)):
        if a[i]==-1:
            break
        if b[i] in hashMapOfa:
            matchingGens[b[i]] = [hashMapOfa[b[i]],i]
        hashMapOfa[a[i]] = i
    #get a random gen and make the crossover
    if len(matchingGens)>0:
        random_gene = choices(list(matchingGens.keys()))[0]
        cut_index = matchingGens[random_gene]
    else:
        cut_index = [0,0]

    offspring_a = a[:cut_index[0]] + b[cut_index[1]:]
    offspring_b = b[:cut_index[1]] + a[cut_index[0]:]

    return offspring_a,offspring_b
```

Another way to promovate the change throughout the generations is the mutation function, in this case the last station is taken and if it mutates it calls the selectRoute method to continue its path.

```python

def mutation(genome: Genome, probability: float = 0.9) -> Genome:

    if genome[1] == -1:
        return genome
    visitedStations = []
    i = 0
    while genome[i+1] != -1 and i<len(genome)-2:
        visitedStations.append(genome[i])
        i+=1
    visitedStations.append(genome[i])

    #get next part of path that will be appended if it mutates
    penultimate_index = genome[i-1]
    toAppendGenome = []
    toAppendGenome = selectRoute(penultimate_index,visitedStations,toAppendGenome,35-penultimate_index)

    # Mutate last element if the probability allows it
    if random() <= probability:
        if i > 0:
          for station in toAppendGenome:
            if i-1<35:
              try:
                #mutate
                genome[i-1] = station
                i+=1
              except:
                """
                error
                """
    #ensure the size of the genome
    while len(genome)>35:
        genome.pop()
    while len(genome) < 35:
        genome.append(-1)

    return genome
```

Gets all the sumatory of the fitness of the whole population

```python
def population_fitness(population: Population, fitness_func: FitnessFunc) -> int:
    return sum([fitness_func(genome) for genome in population])
```

Selects two random genomes of the population that will be parents for next generation.

```python
#this function will allow to select two random parents for a genome for the next generation
def selection_pair(population: Population, fitness_func: FitnessFunc) -> Population:
    fitness_values = [fitness_func(gene) for gene in population]
    inverted_fitness_values = [1 / fitness if fitness != 0 else float('inf') for fitness in fitness_values]
    weights = inverted_fitness_values

    return choices(
        population=population,
        weights=weights,
        k=2
    )
```

Sorts the population by which one is the more fit

```python
def sort_population(population: Population, fitness_func: FitnessFunc) -> Population:
    return sorted(population, key=fitness_func, reverse=True)
```

## Results
This part of the code will present the calling of all the functions in a logical way to get the GA working.

```python
def genome_to_string(genome: Genome) -> str:
    return "".join(map(str, genome))


def print_stats(population: Population, generation_id: int, fitness_func: FitnessFunc):
    print("GENERATION %02d" % generation_id)
    print("=============")
    print("Population: [%s]" % ", ".join([genome_to_string(gene) for gene in population]))
    print("Avg. Fitness: %f" % (population_fitness(population, fitness_func) / len(population)))
    sorted_population = sort_population(population, fitness_func)
    print(
        "Best: %s (%f)" % (genome_to_string(sorted_population[0]), fitness_func(sorted_population[0])))
    print("Worst: %s (%f)" % (genome_to_string(sorted_population[-1]),
                              fitness_func(sorted_population[-1])))
    print("")

    return sorted_population[0]


def run_evolution(
        populate_func: PopulateFunc,
        fitness_func: FitnessFunc,
        fitness_limit: int,
        selection_func: SelectionFunc = selection_pair,
        crossover_func: CrossoverFunc = single_point_crossover,
        mutation_func: MutationFunc = mutation,
        generation_limit: int = 100,
        printer: Optional[PrinterFunc] = None) \
        -> Tuple[Population, int]:
    population = populate_func()
    #for each generation do the following
    for i in range(generation_limit):
        #get the population in order based in the fittness
        population = sorted(population, key=lambda genome: fitness_func(genome), reverse=False)
        #check ties in first place
        firstNotTied = False
        i = 0
        tiedGenes = [population[0]]
        while not firstNotTied and i<len(population)-1:
          if fitness_func(population[i])==fitness_func(population[i+1]):
            tiedGenes.append(population[i+1])
            i+=1
          else:
            firstNotTied = True
        if len(tiedGenes)>1:
            firstPlace,secondPlace = find_minimum_weight_genome(tiedGenes)
            next_generation = [firstPlace,secondPlace]
        else:
            #check ties in second place
            next_generation = [population[0]]
            firstNotTied = False
            i = 1
            tiedGenes = [population[1]]
            while not firstNotTied and i<len(population)-1:
              if fitness_func(population[i])==fitness_func(population[i+1]):
                tiedGenes.append(population[i+1])
                i+=1
              else:
                firstNotTied = True
            if len(tiedGenes)>1:
                firstPlace,secondPlace = find_minimum_weight_genome(tiedGenes)
                next_generation.append(firstPlace)
            else:
                next_generation = population[1]


        if printer is not None:
            printer(population, i, fitness_func)
        #if arrive to the goal finish
        if fitness_func(population[0]) == fitness_limit:
            break
        #elitism
        next_generation = population[0:2]
        #complete next gen
        for j in range(int(len(population) / 2) - 1):
            parents = selection_func(population, fitness_func)
            offspring_a, offspring_b = crossover_func(parents[0], parents[1])
            offspring_a = mutation_func(offspring_a)
            offspring_b = mutation_func(offspring_b)
            next_generation += [offspring_a, offspring_b]

        population = next_generation

    return population, i
```
